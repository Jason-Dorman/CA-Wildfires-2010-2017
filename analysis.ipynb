```python
import pandas as pd
```

```python
fires = pd.read_csv('all_fires.csv')

fires.head()
```

```python
fires.dtypes
```

```python
totals = fires['Total'].sum()
arson = fires['Arson'].sum()
campfires = fires['Campfire'].sum()
debris = fires['Debris Burning'].sum()
electrical = fires['Elec. Power'].sum()
equipment = fires['Equip. Use'].sum()
lightning = fires['Ltng.'].sum()
misc = fires['Misc.'].sum()
pwf = fires['P-W-F'].sum()
rails = fires['Railroad'].sum()
smoking = fires['Smoking'].sum()
vehicle = fires['Vehicle'].sum()
undetermined = fires['Undet.'].sum()

print('Total: ', totals)
print('Arson: ', arson)
print('Campfires: ', campfires)
print('Burning Debris: ', debris)
print('Electrical: ', electrical)
print('Equipment: ', equipment)
print('Lightning: ', lightning)
print('Misc.: ', misc)
print('PWF: ', pwf)
print('Railroads: ', rails)
print('Smoking: ', smoking)
print('Vehicle: ', vehicle)
print('Undetermined: ', undetermined)
```

```python
arson_p = (arson / totals) * 100
campfires_p = (campfires / totals) * 100
debris_p = (debris / totals) * 100
electrical_p = (electrical / totals) * 100
equipment_p = (equipment / totals) * 100
lightning_p = (lightning / totals) *100
misc_p = (misc / totals) * 100
pwf_p = (pwf / totals) * 100
rails_p = (rails / totals) * 100
smoking_p = (smoking / totals) *100
vehicle_p = (vehicle / totals) *100
undetermined_p = (undetermined / totals) * 100

print('Arson Percentage: ', arson_p)
print('Campfire Percentage: ', campfires_p)
print('Debris Percentage: ', debris_p)
print('Electrical Percentage: ', electrical_p)
print('Equipment Percentage: ', equipment_p)
print('Lightning Percentage: ', lightning_p)
print('Misc. Percentage: ', misc_p)
print('PWF Percentage: ', pwf_p)
print('Railroad Percentage: ', rails_p)
print('Smoking Percentage: ', smoking_p)
print('Vehicle Percentage: ', vehicle_p)
print('Undetermined Percentage: ', undetermined_p)
```
